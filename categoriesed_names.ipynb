```python
import pandas as pd
import random as rd
```

```python
file= pd.read_excel('names_Db.xlsx')
```

```python
file
```

```python
def _rand_name(array):
    choice = rd.choice(array)
    fname, lname, rel = choice
    while True:
        lc = rd.choice(array)
        if lc[2] == rel:
#             print(lc[2], rel)
            lname = lc[1]
            break
    return [fname, lname, rel]
```

```python
def name_generator(array,count, fluxing=True):
    var = []
    if not fluxing:
        var = array
    else:
        for c in range(count):
            var.append(_rand_name(array))
    rd.shuffle(var)
    return pd.DataFrame(var, columns=['first_name', 'last_name', 'rel'])
```

```python
### Call name generator function along with count of names wanted for this operation
names = name_generator(file.values, 102)
names.to_csv('names.csv')
```
